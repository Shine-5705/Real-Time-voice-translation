```python
import re
from wordfreq import zipf_frequency

ENGLISH_THRESHOLD = 0.90

def english_word_metrics(text: str) -> tuple[int, int, float]:
    tokens = re.findall(r"[A-Za-z']+", str(text).lower())
    total_words = len(tokens)
    if total_words == 0:
        return 0, 0, 0.0

    english_words = sum(1 for token in tokens if zipf_frequency(token, "en") > 0)
    ratio = english_words / total_words
    return english_words, total_words, ratio

metrics = df2['source_text'].fillna('').apply(english_word_metrics)

df2[['english_words_detected', 'total_words', 'english_word_ratio']] = pd.DataFrame(
    metrics.tolist(),
    index=df2.index,
)

df2['is_english_by_vocab'] = df2['english_word_ratio'] > ENGLISH_THRESHOLD
df2['asr_decision'] = df2['is_english_by_vocab'].map({
    True: 'accept',
    False: 'reject',
})

sort_col = 'created_at' if 'created_at' in df2.columns else 'timestamp'
sorted_df2 = df2.sort_values(by=sort_col, ascending=False)

sorted_df2[[
    'session_id',
    'source_text',
    'english_words_detected',
    'total_words',
    'english_word_ratio',
    'asr_decision',
]].head(20)
```

```python

import base64
import io
import wave
import pandas as pd
import numpy as np

def wav_duration_from_b64(s: str):
    if pd.isna(s) or not s:
        return None
    try:
        # handle possible data-URL prefix
        if "," in s and s.strip().startswith("data:"):
            s = s.split(",", 1)[1]

        audio_bytes = base64.b64decode(s)
        with wave.open(io.BytesIO(audio_bytes), "rb") as wf:
            n_frames = wf.getnframes()
            sr = wf.getframerate()
            return n_frames / float(sr) if sr else None
    except Exception:
        return None

final_df["audio_duration_s"] = final_df["audio_base64"].apply(wav_duration_from_b64)
```

```python
def compile_audio(df: pd.DataFrame, output_path: str, sample_rate: int = 22050, audio_base64_col: str = "audio_base64"):
    df = df.copy().reset_index(drop=True)
    df["desired_start_s"] = df["relative_time_s"] + df["total_latency"]

    actual_starts = []
    for i in range(len(df)):
        desired = df.loc[i, "desired_start_s"]
        if actual_starts:
            prev_end = actual_starts[-1] + df.loc[i - 1, "audio_duration_s"]
            start = max(desired, prev_end)
        else:
            start = desired
        actual_starts.append(start)

    df["actual_start_s"] = actual_starts
    df["actual_end_s"] = df["actual_start_s"] + df["audio_duration_s"]

    total_samples = int(np.ceil(df["actual_end_s"].max() * sample_rate))
    output = np.zeros(total_samples, dtype=np.int16)

    for _, row in df.iterrows():
        audio_bytes = base64.b64decode(row[audio_base64_col])
        with wave.open(io.BytesIO(audio_bytes), "rb") as wf:
            samples = np.frombuffer(wf.readframes(wf.getnframes()), dtype=np.int16)
        start = int(row["actual_start_s"] * sample_rate)
        end = min(start + len(samples), total_samples)
        output[start:end] = samples[: end - start]

    with wave.open(output_path, "wb") as wf:
        wf.setnchannels(1)
        wf.setsampwidth(2)
        wf.setframerate(sample_rate)
        wf.writeframes(output.tobytes())

    print(f"Saved {output_path} — {df['actual_end_s'].max():.2f}s, {total_samples} samples")
    return df[["desired_start_s", "actual_start_s", "audio_duration_s", "actual_end_s"]]
```

```python
import re
from datetime import datetime

def parse_transcript_segments(log_text):
    lines = [l.strip() for l in log_text.strip().split("\n") if l.strip()]
    
    pattern = re.compile(
        r'^(\d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2}\.\d+)\s*\|.*?\[SRT\]\s*(Interim|Final)\s+Speaker=(\S+)\s*\|\s*(.*)$'
    )
    
    segments = []
    current_interims = []
    segment_start_time = None
    
    for line in lines:
        m = pattern.match(line)
        if not m:
            continue
        
        timestamp_str, result_type, speaker, text = m.groups()
        
        if result_type == "Interim":
            if segment_start_time is None:
                segment_start_time = timestamp_str
            current_interims.append(text)
        
        elif result_type == "Final":
            segment_end_time = timestamp_str
            if segment_start_time is None:
                segment_start_time = timestamp_str
            
            segments.append({
                "segment_start_time": segment_start_time,
                "segment_end_time": segment_end_time,
                "interim_results": current_interims,
                "final_result": text,
                "speaker": speaker,
            })
            current_interims = []
            segment_start_time = None
    
    # handle trailing interims with no final
    if current_interims and segment_start_time:
        segments.append({
            "segment_start_time": segment_start_time,
            "segment_end_time": timestamp_str,
            "interim_results": current_interims,
            "final_result": None,
            "speaker": None,
        })
    
    return segments

segments = parse_transcript_segments(transcripts_log)
print(f"Total segments: {len(segments)}")
print()
for i, seg in enumerate(segments[:5]):
    print(f"--- Segment {i} ---")
    print(f"  Start: {seg['segment_start_time']}")
    print(f"  End:   {seg['segment_end_time']}")
    print(f"  Speaker: {seg['speaker']}")
    print(f"  Interims ({len(seg['interim_results'])}): [{seg['interim_results'][0]!r} ... {seg['interim_results'][-1]!r}]" if seg['interim_results'] else "  Interims: []")
    print(f"  Final: {seg['final_result']!r}")
    print()
```

```python
from datetime import timedelta

segment_lookup = {seg['final_result']: seg for seg in segments}

ist_offset = timedelta(hours=5, minutes=30)

def lookup_segment(source_text):
    seg = segment_lookup.get(source_text)
    if seg is None:
        return pd.Series({
            'segment_start_time_ist': None,
            'segment_end_time_ist': None,
            'segment_duration_s': None,
            'num_interims': None,
        })
    
    start = datetime.strptime(seg['segment_start_time'], '%Y-%m-%d %H:%M:%S.%f') + ist_offset
    end = datetime.strptime(seg['segment_end_time'], '%Y-%m-%d %H:%M:%S.%f') + ist_offset
    duration = (end - start).total_seconds()
    
    return pd.Series({
        'segment_start_time_ist': start,
        'segment_end_time_ist': end,
        'segment_duration_s': round(duration, 3),
        'num_interims': len(seg['interim_results']),
    })

new_cols = ['segment_start_time_ist', 'segment_end_time_ist', 'segment_duration_s', 'num_interims']
dupe_cols = [c for c in final_df.columns if c in new_cols or any(c.startswith(n + '.') for n in new_cols)]
if dupe_cols:
    final_df = final_df.drop(columns=dupe_cols)

timing_cols = final_df['source_text'].apply(lookup_segment)
final_df[new_cols] = timing_cols
final_df['segment_start_time_ist'] = pd.to_datetime(final_df['segment_start_time_ist'])
final_df['segment_end_time_ist'] = pd.to_datetime(final_df['segment_end_time_ist'])

matched = final_df['segment_start_time_ist'].notna().sum()
print(f"Matched {matched}/{len(final_df)} rows")
final_df[['source_text', 'segment_start_time_ist', 'segment_end_time_ist', 'segment_duration_s', 'num_interims']].head(10)
```

```python
import aiohttp
import asyncio

BASE_URL = "http://172.16.9.11:10021"
TRANSLATION_PATH = "/voice_stream/translation"

async def translate_hi_to_en(text, session):
    payload = {
        "translation_id": "trans_001",
        "secret_key": "gnani_translate_2026",
        "source_lang": "hi-IN",
        "target_lang": "en-IN",
        "text": text,
    }
    async with session.post(
        f"{BASE_URL}{TRANSLATION_PATH}",
        json=payload,
        headers={"Content-Type": "application/json"},
    ) as response:
        response.raise_for_status()
        return await response.json()

async def translate_all(texts):
    async with aiohttp.ClientSession() as session:
        results = []
        for i, text in enumerate(texts):
            try:
                result = await translate_hi_to_en(text, session)
                translated = result if isinstance(result, str) else result.get("translated_text", result.get("text", str(result)))
                results.append(translated)
            except Exception as e:
                print(f"[{i}] Error: {e}")
                results.append(None)
            if (i + 1) % 20 == 0:
                print(f"Translated {i + 1}/{len(texts)}")
        return results

texts = final_df['source_text'].tolist()
english_texts = await translate_all(texts)
final_df['english_text'] = english_texts

print(f"Done. Translated {sum(1 for t in english_texts if t is not None)}/{len(texts)}")
final_df[['source_text', 'english_text']].head(10)
```
