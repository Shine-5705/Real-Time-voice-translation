```python
from requests import Session
import json
import time
from uuid import uuid4
```

```python
session = Session()
```

```python
domain = 'https://orchestrator.inya.ai'
flow_id = 'prod-d913d567cba5'

_id = str(uuid4())

def post_initial_message(_id):
    url = f'{domain}/initial_message?call_id={_id}&flow_id={flow_id}&environment=production&organization_id=common&mobile=7908193750&user_id=f5e71f9c-b89b-4fd4-b2be-e00302172865'

    response = session.get(url)
    response.json()
    
```

```python
times = {}

def send_voice_stream(_id,text):
    url = f'{domain}/voice_stream'

    data = {
        "user_interrupted": False,
        "call_id": _id,
        "flow_id": flow_id,
        "total_results": {},
        "asr_transcript_detected": "",
        "detected_language": "en-IN",
        "phone_number": "7908193750",
        "asr_response": text,
        "bot_language": "en-IN"
    }

    start_time = time.time()
    response = session.post(url, json=data, stream=True)
    response.raise_for_status()

    first_chunk_at = None
    events = []

    for line in response.iter_lines(decode_unicode=True):
        if not line:
            continue
        payload = json.loads(line)
        if payload['nlp_response'] == 'EOT':
            continue
        if first_chunk_at is None:
            first_chunk_at = time.time()
        events.append(payload)
        print(payload)

    end_time = time.time()

    if first_chunk_at:
        print(f"TTFB: {first_chunk_at - start_time:.3f}s")
        if _id not in times:
            times[_id] = []
        times[_id].append(round(first_chunk_at - start_time, 3))

    print(f"Total stream time: {end_time - start_time:.3f}s")
    print(f"Events received: {len(events)}")
```
